```python
import pandas as pd #Manipulação e análise de dados, fornece DataFrames e Series.
import numpy as np #Computação numérica, operações com arrays e álgebra linear
import matplotlib.pyplot as plt #Criação de gráficos e visualizações.
import seaborn as sns #Visualizações estatísticas, construído sobre o matplotlib.
```

# Plotando um gráfico, que mostra dia a dia a taxa de desemprego.

```python
# Visualização do gráfico de desemprego pelo ano entre homens e mulheres.

plt.figure(figsize = (10,6)) # tamanho do gráfico
plt.plot(df["date"], df["unrate_men"], color = "blue", label = "Homens") #plotar dados da coluna data x taxa homens
plt.plot(df["date"],df["unrate_women"], color = "red", label = "Mulheres") #plotar dados da coluna data x taxa mulheres
plt.title("Taxa de desemprego entre gêneros") # Titulo
plt.xlabel("Ano") # legenda do eixo X
plt.ylabel("Taxa de desemprego(%)") # legenda do eixo y
plt.legend() # chamar a legenda (label= "Homens" e "mulheres")
plt.grid() # criar grades
plt.show() # iniciar plotagem
```

Foi possivel observar que tivemos o maior taxa de desemprego já registada no ano de 2020, devido a pandemia do covid-19.

Tivemos outros dois periodos onde as taxas foram muito altas para os padrões dos EUA, altas que se mostrará muito maior no gênero masculino.Podemos levar em consideração que o gênero masculino seja maior no mercado de trabalho, ocasionando assim um maior número de desempregados nos momentos de crise.

```python
# Convertendo a coluna 'date' para o tipo datetime
df['date'] = pd.to_datetime(df['date'])

# Agrupando os dados por ano e calculando a média da taxa de desemprego por gênero
df_anual = df.groupby(df['date'].dt.year)[['unrate_men', 'unrate_women']].mean().reset_index()

# Criando o gráfico de linha
plt.figure(figsize=(10, 6))
plt.plot(df_anual['date'], df_anual['unrate_men'], marker='o', label='Homens')
plt.plot(df_anual['date'], df_anual['unrate_women'], marker='o', label='Mulheres')
plt.xlabel('Ano')
plt.ylabel('Taxa de Desemprego')
plt.title('Taxa de Desemprego por Gênero e Ano')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculando a correlação de Pearson
correlacao = df_anual['unrate_men'].corr(df_anual['unrate_women'])

print("Correlação de Pearson entre as taxas de desemprego de homens e mulheres:", correlacao)
```

```python
# Selecionando apenas as colunas de interesse
df_faixas_etarias = df[['unrate_16_to_17', 'unrate_18_to_19', 'unrate_20_to_24', 'unrate_25_to_34',
                       'unrate_35_to_44', 'unrate_45_to_54', 'unrate_55_over']]

# Calculando a matriz de correlação
correlacao = df_faixas_etarias.corr()

# Visualizando a matriz de correlação usando um heatmap
sns.heatmap(correlacao, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação entre as Faixas Etárias')
plt.show()
```

Podemos observar que a correlação mais forte entre faixa de idades é entre 16 para 17 anos com a faixa etaria de 18 para 19 anos; Já a mais fraca esta entre a faixa etaria 16 para 17 anos com a faixa de 55 anos ou mais.

Esta situação pode indicar que tem uma quantidade muito maior de pessoas no mercado de trabalho com idades entre 16 e 17 anos e uma quantidade muito menor de pessoas com 55 anos ou mais.

Podendo assim ser possivel levatar uma hipotese, de que as pessoas de 55 anos ou mais, tenham mais estabilidade em seus empregos.
